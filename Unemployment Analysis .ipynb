```python
df.columns = ["States" , "Date" , "Frequency" , "Unemployment Rate" , "Employed", "Labour Participation" , "Area"]
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(4, 3))
sns.heatmap(df.corr(), cmap = "Blues")
plt.show()
```

```python
#Area wise estimated employeed
df.columns= ["States","Date","Frequency",
               "Unemployment Rate","Employed",
               "Labour Participation Rate","Area"]
plt.title("Indian Unemployment")
sns.histplot(x="Employed", hue="Area", data=df)
plt.show()
```

```python
#State wise unemployment rate 
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Unemployment Rate", hue="States", data=df)

plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 5))
sns.heatmap(data=df[['Unemployment Rate']], annot=True)
plt.show()
```
